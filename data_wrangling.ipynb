```python
import pandas as pd

# Load the exported raw file (ignore the first few non-genotype columns)
raw_file = './data_/genotype_sequences.raw'
df = pd.read_csv(raw_file, delim_whitespace=True)

# Drop non-genotype columns (FID, IID, etc.)
genotype_df = df.iloc[:, 6:]

# Convert allele pairs to a single consensus base (simplifying for this example)
```

```python
genotype_df
```

```python
def get_consensus(allele1, allele2):
    if allele1 == allele2:
        return allele1  # Homozygous
    else:
        return allele1  # Assume first allele for heterozygous (or can randomize)

# Convert genotype columns to a sequence string
sequence = ''.join(genotype_df.apply(lambda x: str(get_consensus(x[0], x[1])), axis=1))

# Save sequence in FASTA format
with open('./data_/genotype_sequences.fasta', 'w') as fasta_file:
    fasta_file.write(">sample_1\n")
    fasta_file.write(sequence + "\n")

print("FASTA file generated successfully!")
```

```python
# Define reference and alternate alleles for each SNP (mock example, update as needed)
reference_alleles = ['A', 'C', 'G', 'T', 'A', 'C', 'G', 'T']  # Replace with actual reference alleles
alternate_alleles = ['G', 'T', 'A', 'C', 'G', 'T', 'A', 'C']  # Replace with actual alternate alleles

# Function to convert genotype to nucleotide sequence
def genotype_to_nucleotide(genotype, ref, alt):
    if genotype == 0:
        return ref  # Homozygous reference
    elif genotype == 1:
        return ref  # Heterozygous (choose ref, can be adjusted)
    elif genotype == 2:
        return alt  # Homozygous alternate
    else:
        return 'N'  # Missing or unknown genotype
# Convert each SNP genotype to a nucleotide sequence
sequence_list = []
for i, col in enumerate(genotype_df.columns):
    ref_allele = reference_alleles[i % len(reference_alleles)]  # Use actual SNP data
    alt_allele = alternate_alleles[i % len(alternate_alleles)]
    sequence_list.append(genotype_df[col].apply(lambda x: genotype_to_nucleotide(x, ref_allele, alt_allele)))

# Concatenate all SNPs into a DNA sequence
sequence = ''.join(sequence_list[i].astype(str).values[0] for i in range(len(sequence_list)))
```

```python
# Save sequence in FASTA format
with open('./data_/genotype_sequences.fasta', 'w') as fasta_file:
    fasta_file.write(">sample_1\n")
    fasta_file.write(sequence + "\n")

print("FASTA file generated successfully!")
```
