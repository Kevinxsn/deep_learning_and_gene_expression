```python
import pandas as pd
import numpy as np
```

```python
range_data = pd.read_csv('../../data/snp_sequence_ranges.txt', delimiter='\t')
range_data
```

```python
chr1_reulst = pd.read_csv('/projects/ps-renlab2/sux002/DSC180/local_testing/result/chr1_numer_1000_result.csv', delimiter='\t')
```

```python
chr1_reulst = chr1_reulst.sort_values(['Result'], ascending=False)
chr1_reulst.head()
```

```python
chr1_reulst['abs_result'] = abs(chr1_reulst['Result'])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(chr1_reulst['Position'], chr1_reulst['abs_result'], alpha=0.7, edgecolor='k')


plt.title('Dot Plot of SNP Positions vs. Result Values', fontsize=14)
plt.xlabel('Position (SNP)', fontsize=12)
plt.ylabel('Difference (Result)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
