```python
# Load reference genome (hg38)
ref_genome = Fasta("38.fa")

# Load VCF file (genotyped variants)
vcf_reader = VCF("./data_/genotype_sequences2.vcf")
```

```python
# Function to reconstruct sequence with genotype data
def reconstruct_sequence(chrom, start, end):
    """Generate a DNA sequence from the reference genome and incorporate VCF genotypes."""
    # Extract reference sequence from hg38
    seq = list(ref_genome[chrom][start:end].seq)  # Convert to list for mutability

    for record in vcf_reader:
        pos = record.POS - start  # Adjust position to local index
        if start <= record.POS < end:  # Ensure variant is within the requested range
            genotype = record.genotypes[0][:2]  # Get first sample's genotype

            # If heterozygous (0/1), replace with ALT allele
            if genotype == [0, 1]:
                seq[pos] = record.ALT[0]

            # If homozygous alternate (1/1), fully replace with ALT allele
            elif genotype == [1, 1]:
                seq[pos] = record.ALT[0]

            # If homozygous reference (0/0), keep reference base

    return "".join(seq)
```

```python
# Get the first variant's position
variant_positions = [record.POS for record in vcf_reader]
start = variant_positions[0]
end = start + 1000  # Extract 1kb around it

print(f"Selected region: chr22:{start}-{end}")
```

```python
# Define chromosome, start, and end position (adjust as needed)
chrom = "22"
start, end = 16390411, 50780578  # Example range

# Get modified DNA sequence
dna_sequence = reconstruct_sequence(chrom, start, end)

# Output the reconstructed sequence
```

```python
# Save sequence in FASTA format
with open('./data_/chr22_dnaseq.fasta', 'w') as fasta_file:
    fasta_file.write(f">chr22:{start}-{end}\n")
    fasta_file.write(dna_sequence + "\n")

print("FASTA file generated successfully!")
```
